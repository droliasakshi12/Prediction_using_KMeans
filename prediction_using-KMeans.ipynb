```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#creating sample dataset
data = {
    'customer':['riya','aman','faizan','neha','imran','sneha'],
    'age': [20,30,40,22,38,25],
    'spending':[100,200,300,110,290,130]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
x = df[['age','spending']]
x
```

```python
#loading a model 
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2 , random_state = 42 , n_init = 10)

df['group'] = model.fit_predict(x)
plt.figure(figsize = (6,5))


#using unique function to remove the repeated values  
for group in df['group'].unique():
    group_data = df[df['group'] == group]  

    #used df[df['group]== group] outside because the value provided will be in a list so it will convert it in to rows 
    plt.scatter(group_data['age'] , group_data['spending'] , label=f'group {group}')
    print(group_data)


plt.title('Customer Segement (K-Means)')
plt.xlabel("age")
plt.ylabel("spending")
plt.legend()
plt.grid(True)
plt.show()
```
